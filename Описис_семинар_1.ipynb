<h1>Построение модели функции синуса с заданным количеством шагов дискретизации</h1>

<h3>Введем необходимые библиотеки для отображения и построения функции синуса</h3>

```python
import matplotlib.pyplot as plt
import numpy as np
```

<h3>Напишем функции для дискретизации синуса</h3>

```python
def discr_sin(f, T, n):
  x = np.linspace(0, T, n)
  y = np.sin(f*x)
  return (x, y)
```

<h3>Напишем функцию для визуализации полученной функции<h3>

```python
def plot_out(func):
  plt.plot(*func)
```

<h3>Ниже приведен график для отображения синуса через введенные опорную частоту, временной интервал и число временных отсчетов.</h3>
<h5>Входные данные:<br>f=1<br>T=20<br>n=20</h5>

```python
f = float(input())
T = float(input())
n = int(input())
print(f'f={f}, T={T}, n={n}')
func = discr_sin(f, T, n)
plot_out(func)
```
